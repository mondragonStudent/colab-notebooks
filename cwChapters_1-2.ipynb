```
"""
Name: Jose Mondragon
Date: January 15, 2026
Program Name: Jose's Multiplier - cwChapters_1-2.py
Description: This program ask the user to enter a number and multiplies it
by the number of letters in my first name (Jose, which is 4).
"""

# Prompt the user to enter a number
user_input = input("Please enter a number: ")

# Convert the input to a float to allow for decimal calculations
number = float(user_input)

# "Jose" has 4 letters
NAME_LENGTH = 4

# Perform the calculation
calculated_result = number * NAME_LENGTH

# Display the final result
print(f"The number {number} multiplied by {NAME_LENGTH} (letters in 'Jose') is: {calculated_result}")
```
