```
# Hide Drive notifications
from google.colab import output
output.eval_js('''
    try {
        // Access the main Colab document (parent window)
        const targetDoc = window.parent.document;

        // Create a MutationObserver to watch the main window
        const observer = new MutationObserver((mutations) => {
            const notification = targetDoc.querySelector('#error-dialog');
            if (notification) {
                notification.style.display = "none";
            }
        });

        // Start observing the body of the main window
        observer.observe(targetDoc.body, { childList: true, subtree: true });

        // Also try to hide immediately
        const notification = targetDoc.querySelector('#error-dialog');
        if (notification) {
            notification.style.display = "none";
        }
    } catch (e) {
        console.error("Could not access parent document. Ensure you are running this in a standard Colab environment.", e);
    }
''')
```
