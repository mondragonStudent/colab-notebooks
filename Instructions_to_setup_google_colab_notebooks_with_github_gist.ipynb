# **How to Use Google Colab with Your MDC Account: Accessing Google Colab Pro without Google Drive**
<small>Created by student Jose Mondragon</small><br><br>


**The Challenge & The Solution**<br><br>
Although MDC students and teachers receive free access to Google Colab Pro, the college restricts access to Google Drive via student emails. This prevents Colab from saving notebooks directly to Drive.

The solution is to use GitHub and GitHub Gists. This is a fully supported feature of Google Colab that allows you to save your work securely using your @mymdc.net email address.

**Part 1: Initial Setup (Do this once)**
**1.** Create a GitHub Account To keep your academic work organized, I recommend creating a new GitHub account using your MDC email address.

>Go to github.com and sign up.

>Note: If you already have a personal GitHub account you prefer to use, you may use that instead.

**2.** Create a Repository for your Notebooks You need a "home" for your initial templates.

>Log in to your GitHub account.

>Click the + icon in the top right corner and select New repository.

>Name it colab-notebooks (or something similar).

>Make sure it is set to Public (or Private if you prefer, but Public is easier for sharing with teachers).

>Click Create repository.

>Example: https://github.com/YourUsername/colab-notebooks

**3.** Add the Template File You need a base file to start your assignments.

>Go to the template location: https://github.com/mondragonStudent/colab-notebooks.

>Click on template.ipynb.

>Download the file to your computer (Click the "Download raw file" button or "Raw" and save as).

>Go back to **YOUR NEW** GitHub repository (colab-notebooks).

>Click Add file > Upload files.

# **Part 2: Working on Assignments**
1. Open the Template in Google Colab

>Go to Google Colab.

>Go to File > Open notebook.
image.png

>Select the GitHub tab in the popup window.

>Enter your GitHub URL (or authorize your GitHub account if asked).

>Select your repository and click on template.ipynb.
image.png

**2.** Save Your Work as a Gist (Crucial Step) Since you cannot save to Drive, and you don't want to overwrite your template constantly, you will save your specific assignment as a "Gist."

>Once the notebook is open, immediately go to File > Save a copy as a GitHub Gist.

>Authorize GitHub if prompted.
image.png

**3.** Rename Your Assignment Google Colab acts as an editor, but the file "lives" in GitHub Gist. To rename it properly:

>In Colab, click the title at the top (e.g., template.ipynb) and rename it to your assignment name (e.g., My_First_Colab.ipynb).

>Go to File > Save (this saves the content).

>Important: To update the filename permanently in your Gist list:

>Go to gist.github.com and open your file.

>Click the Edit button (top right).

>Change the filename at the top of the editor (see the filename box).

>Click Update public gist.

# **Part 3: Resuming Your Work**
When you want to continue working on a previous assignment:

>Go to gist.github.com/yourAccount.

>Click on the notebook you want to work on.

>Click the Open in Colab badge (usually appears at the top of the notebook) OR copy the URL, go to Google Colab, and paste it into the "Open from GitHub" tab.

image.png

**Pro Tips for Students**
>Do not forget to save after every change, just File > Save

>Authorize Once: The first time you connect Colab to GitHub, check the box to include "Private repositories and Gists" so you don't have to re-authorize constantly.

>Use Text Cells: As shown in the template, always use + Text cells to add headings and comments to your code so your professor can understand your logic.
